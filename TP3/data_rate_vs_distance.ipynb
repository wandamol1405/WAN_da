## Relación Date Rate vs Distance para protocolos inalámbricos

Los valores típicos (pueden variar según implementación) son los siguientes.

#### Wi-Fi:

- Distancia: ~10–100 m
- Data rate: hasta ~1 Gbps (Wi-Fi 6)

#### Bluetooth:

- Distancia: ~1–100 m (según clase)
- Data rate: ~1–3 Mbps (BLE: ~2 Mbps)

#### ZigBee:

- Distancia: ~10–100 m
- Data rate: ~250 kbps

#### NFC:

- Distancia: < 0.2 m
- Data rate: ~106 kbps – 424 kbps

#### LTE:

- Distancia: ~1–10 km
- Data rate: hasta ~300 Mbps (LTE-A)

#### GSM:

- Distancia: ~1–10 km
- Data rate: ~9.6 kbps – 100 kbps (EDGE hasta ~384 kbps)

#### 5G (3GPP):

- Distancia: ~1–10 km (depende de banda)
- Data rate: hasta ~10 Gbps

#### LoRa:

- Distancia: ~2–15 km
- Data rate: ~0.3 – 50 kbps

#### NB-IoT:

- Distancia: ~1–10 km
- Data rate: ~250 kbps (máx. 1 Mbps uplink/downlink)

#### SigFox:

- Distancia: ~10–40 km (rural)
- Data rate: ~100 bps

#### Z-Wave:

- Distancia: ~30–100 m
- Data rate: ~100 kbps

Mediante el siguiente script generamos el gráfico que relaciona estos valores para cada protocolo:

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Diccionario con protocolos y sus rangos aproximados
protocols = {
    "NFC": (0.01, 0.2, 100e3, 500e3),         # 100–500 kbps, <0.2 m
    "Bluetooth": (1, 100, 1e6, 3e6),          # 1–3 Mbps, hasta 100 m
    "ZigBee": (1, 100, 20e3, 250e3),          # 20–250 kbps, hasta 100 m
    "Z-Wave": (1, 100, 10e3, 100e3),          # 10–100 kbps, hasta 100 m
    "Wi-Fi": (1, 100, 10e6, 1e9),             # 10 Mbps – 1 Gbps, hasta 100 m
    "LTE": (100, 10000, 10e6, 1e8),           # 10 Mbps – 100 Mbps, hasta 10 km
    "5G": (100, 10000, 1e9, 10e9),            # 1–10 Gbps, hasta 10 km
    "NB-IoT": (100, 10000, 20e3, 250e3),      # 20–250 kbps, hasta 10 km
    "GSM": (100, 10000, 10e3, 100e3),         # 10–100 kbps, hasta 10 km
    "LoRa": (100, 15000, 0.3e3, 50e3),        # 0.3–50 kbps, hasta 15 km
    "SigFox": (1000, 50000, 100, 1000),       # 100–1000 bps, hasta 50 km
}

# Crear figura
fig, ax = plt.subplots(figsize=(10, 7))

# Usar un mapa de colores
colors = cm.get_cmap("tab10", len(protocols))

# Dibujar rectángulos con colores distintos
for i, (name, (d_min, d_max, r_min, r_max)) in enumerate(protocols.items()):
    ax.add_patch(plt.Rectangle((d_min, r_min),
                               d_max - d_min,
                               r_max - r_min,
                               alpha=0.4,
                               color=colors(i),
                               label=name))

# Escalas logarítmicas
ax.set_xscale("log")
ax.set_yscale("log")

# Etiquetas y título
ax.set_xlabel("Distancia (m)")
ax.set_ylabel("Data Rate (bps)")
ax.set_title("Protocolos: Data rate vs Distancia (Rangos aproximados)")

# Leyenda
ax.legend(loc="best")

# Mostrar gráfico
plt.show()
```

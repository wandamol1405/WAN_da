## Relación Date Rate vs Distance para protocolos inalámbricos

Los valores típicos (pueden variar según implementación) son los siguientes.

#### Wi-Fi:

- Distancia: ~10–100 m

- Data rate: hasta ~1 Gbps (Wi-Fi 6)

#### Bluetooth:

- Distancia: ~1–100 m (según clase)

- Data rate: ~1–3 Mbps (BLE: ~2 Mbps)

#### ZigBee:

- Distancia: ~10–100 m

- Data rate: ~250 kbps

#### NFC:

- Distancia: <10 cm

- Data rate: ~106 kbps – 424 kbps

#### LTE:

- Distancia: ~1–10 km

- Data rate: hasta ~300 Mbps (LTE-A)

#### GSM:

- Distancia: ~1–10 km

- Data rate: ~9.6 kbps – 100 kbps (EDGE hasta ~384 kbps)

#### 5G (3GPP):

- Distancia: ~1–10 km (depende de banda)

- Data rate: hasta ~10 Gbps

#### LoRa:

- Distancia: ~2–15 km

- Data rate: ~0.3 – 50 kbps

#### NB-IoT:

- Distancia: ~1–10 km

- Data rate: ~250 kbps (máx. 1 Mbps uplink/downlink)

#### SigFox:

- Distancia: ~10–40 km (rural)

- Data rate: ~100 bps

#### Z-Wave:

- Distancia: ~30–100 m

- Data rate: ~100 kbps

Mediante el siguiente script generamos el gráfico que relaciona estos valores para cada protocolo:

```python
import matplotlib.pyplot as plt

protocolos = ["Wi-Fi","Bluetooth","ZigBee","NFC","LTE","GSM","5G","LoRa","NB-IoT","SigFox","Z-Wave"]
dist = [50, 10, 50, 0.1, 5000, 5000, 5000, 10000, 5000, 20000, 50]  # metros aprox.
rate = [1e9, 2e6, 2.5e5, 4e5, 3e8, 1e5, 1e10, 5e4, 2.5e5, 100, 1e5]

plt.figure(figsize=(8,6))
plt.xscale("log")
plt.yscale("log")
plt.scatter(dist, rate)

for i, txt in enumerate(protocolos):
    plt.annotate(txt, (dist[i], rate[i]))

plt.xlabel("Distancia [m]")
plt.ylabel("Data rate [bps]")
plt.title("Protocolos: Data rate vs Distancia")
plt.grid(True, which="both")
plt.show()
```
